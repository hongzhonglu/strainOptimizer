```python
df_targets=pd.read_csv(r'analysis_code\results/sclareolCellfac_predict_targts.csv',index_col=0)
df_targets.head()
```

```python
df_data=df_targets['pathway'].value_counts()
# remove transporter and unknown
df_data=df_data[~df_data.index.isin(['Transporter','unknown'])]

fig,ax=plt.subplots(figsize=(1.5,1.8))
sns.barplot(x=df_data.values,y=df_data.index,orient='h',ax=ax)
ax.set_xlabel('Target number',fontsize=8)
ax.set_ylabel('')
# set fontsize for y tick
ax.tick_params(axis='y',labelsize=8)
ax.tick_params(axis='x',labelsize=8)
# ax.set_title('Pathway distribution of targets',fontsize=15)

# save figure
plt.tight_layout()
plt.savefig(r'analysis_code\figures\output\fig5b_pathway_distribution_of_targets.svg',dpi=600,bbox_inches='tight',transparent=True)
plt.show()
```
